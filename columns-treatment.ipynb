```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
from IPython.display import clear_output

clear_output()
```

```python
df = pd.read_pickle('data/saeb_pb_filtered.pkl')
df.head()
```

```python
# selecting categorical columns
#columns_to_filter = ['possui_eletrecidade', 'possui_escrivaninha', 'escolaridade_mae', 'escolaridade_pai', 'responsaveis_leem', 'responsaveis_incentivam_realizacao_licao_casa', 'responsaveis_incentivam_estudos', 'responsaveis_incentivam_comparecer_aulas', 'tempo_chegada_escola', 'forma_chegada_escola', 'transporte_escolar', 'tempo_lazer', 'tempo_cursos', 'tempo_estudos', 'leitura_noticias', 'leitura_livros_geral', 'leitura_historia_quadrinhos', 'sexo', 'escola_publica']
# selecting all columns with dtype object
columns_to_filter = df.select_dtypes(include=['object']).columns

# remove rows with all values '.' or '*'
df = df[(df[columns_to_filter] != '.').all(axis=1)]
df = df[(df[columns_to_filter] != '*').all(axis=1)]
print(len(df))

df.head()
```

```python
# print columns with dtype object
print(df.select_dtypes(include=['object']).columns)
```

```python
df['idade'].value_counts(dropna=False)
```

```python
# nulling '2021' from 'idade'
# not being used as we dont have 2021 data in this dataset
#df['idade'] = df['idade'].apply(lambda x: x if x != 2021 else np.nan)


# change 'idade' type to float (temporarily)
df['idade'] = df['idade'].astype(float)

# fill 'idade' missing values with the mean (int)
df.fillna({'idade': df['idade'].mean()}, inplace=True)
df['idade'] = df['idade'].astype(int)

df['idade'].head()
```

```python
print(df['idade'].value_counts())

print('\n')

print(df['idade'].describe())
```

```python
print(df['proficiencia_saeb'].isnull().sum())

# drop null values from 'proficiencia_saeb'
df.dropna(subset=['proficiencia_saeb'], inplace=True)

print(df['proficiencia_saeb'].isnull().sum())
```

```python
df['desempenho_aluno'].value_counts(dropna=False)
```

```python
# changing 'desempenho_aluno' to categorical numerical values
# Insuficiente = 0, Básico = 1, Proficiente = 2, Avançado = 3
# dropping None values

# drop the ones with value 'None'
_ = df.dropna(subset=['desempenho_aluno']).copy()

_.loc[:, 'desempenho_aluno_cat'] = _['desempenho_aluno'].map({'Insuficiente': 0, 'Básico': 1, 'Proficiente': 2, 'Avançado': 3})
#df['desempenho_aluno'] = df['desempenho_aluno'].astype(int)

df = _

df['desempenho_aluno_cat'].value_counts(dropna=False)
```

```python
for column in columns_to_filter:
    print(column, df[column].unique())
```

```python
# questões "tempo_"
# Fora da escola em dias de aula, quanto tempo você usa para: 
# E: Não uso meu tempo para isso (0)
# A: Menos de 1 hora (1)
# B: Entre 1 a 2 horas (2)
# C: Mais de 2 horas (3)
# D: Mais de 3 horas (4)


tempo_columns = ['tempo_lazer']

for column in tempo_columns:
    df[column] = df[column].map({'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 0})

df[tempo_columns].head()
```

```python
# questões "responsaveis_"
# Com que frequência seus pais ou responsáveis costumam:
# A: Nunca ou Quase Nunca
# B: Raramente
# C: Sempre ou Quase Sempre
# 0, 1, 2

responsaveis = ['responsaveis_incentivam_realizacao_licao_casa', 'responsaveis_incentivam_estudos', 'responsaveis_incentivam_comparecer_aulas', 'responsaveis_incentivam_leitura']

for column in responsaveis:
    df[column] = df[column].map({'A': 0, 'B': 1, 'C': 2})

df[responsaveis].head()
```

```python
df['escolaridade_mae'].value_counts(dropna=False)
```

```python
# questões "escolaridade_" e "forma_chegada_escola"
# usar onehotencoder pois são categóricas nominais

from sklearn.preprocessing import OneHotEncoder

columns_to_onehot = ['escolaridade_mae', 'escolaridade_pai', 'possui_computador']

for column in columns_to_onehot:
    encoder = OneHotEncoder(sparse_output=False)
    encoded = encoder.fit_transform(df[column].values.reshape(-1, 1))
    encoded_df = pd.DataFrame(encoded, columns=[f'{column}_{i}' for i in range(encoded.shape[1])], index=df.index)
    df = pd.concat([df, encoded_df], axis=1)
    df.drop(column, axis=1, inplace=True)

df.head()
```

```python
# colunas 'leitura_'

leitura_columns = [coluna for coluna in df.columns if 'leitura_' in coluna]

for column in leitura_columns:
    df[column] = df[column].map({'A': 0, 'B': 1, 'C': 2})

df[leitura_columns].head()
```

```python
# colunas 'frequenta_'

frequenta_columns = [coluna for coluna in df.columns if 'frequenta_' in coluna]

for column in frequenta_columns:
    df[column] = df[column].map({'A': 0, 'B': 1, 'C': 2})

df[frequenta_columns].head()
```

```python
# label encoder para colunas
# 'faz_licao_casa', 'correcao_licao_casa', 'utiliza_biblioteca_escola'

columns_to_label = ['faz_licao_casa', 'correcao_licao_casa', 'utiliza_biblioteca_escola']

for column in columns_to_label:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])

df[columns_to_label].head()
```

```python
# gosta de estudar disciplina

df['gosta_estudar_disciplina']  = df['gosta_estudar_disciplina'].map({'A': 0, 'B': 1})
```

```python
# label encoder para o resto das colunas
# selecionadas com tipo dtype
exclude = ['sigla_uf', 'id_aluno', 'disciplina', 'desempenho_aluno']

new_columns = df.drop(columns=exclude).select_dtypes(include=['object']).columns

print(new_columns)

for column in new_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])

df[new_columns].head()
```
