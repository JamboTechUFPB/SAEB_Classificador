```python
import pandas as pd
import pickle
```

```python
df = pd.read_pickle('data/saeb_pb_2015.pkl')
print(len(df))
df.head()
```

```python
df1 = pd.read_pickle('data/saeb_pb_2017.pkl')
print(len(df1))
df1.head()
```

```python
print(df1.dropna(axis=1, how='all').columns)
df1 = df1.dropna(axis=1, how='all')
```

```python
print(df.dropna(axis=1, how='all').columns)
df = df.dropna(axis=1, how='all')
```

```python
# read colunas_final to get the columns to use
with open('data/colunas_final.txt', 'r') as f:
    colunas = f.read().splitlines()

# only use columns if they are present in dataframe
colunas_df1 = [col for col in colunas if col in df.columns]
colunas_df2 = [col for col in colunas if col in df1.columns]

df = df[colunas_df1]
df1 = df1[colunas_df2]
```

```python
print(len(df1.columns))
print(len(df.columns))
```

```python
print(df1.columns)
```

```python
print(df.columns)
```

```python
# merge dfs
df = pd.concat([df, df1], axis=0)
print(len(df))
df.head()
```

```python
df.info()
```

```python
save = False
if save:
    df.to_pickle('data/saeb_pb_filtered.pkl')
```

```python
# save first row as dictionary to file
with open('data/random_row.json', 'w') as f:
    # transform 'disciplina' column to index
    _ = df.set_index('disciplina')
    f.write(_[_['id_aluno'] == _.sample()['id_aluno'].values[0]].to_json(orient='index'))
```
