```python
import pandas as pd
import pickle
```

```python
df = pd.read_pickle('data/saeb_pb.pkl')
df.head()
```

```python
with open('data/colunas.txt', 'w') as f:
    f.write('\n'.join(df.columns))
```

```python
columns_to_use = [0, 1, 14, 9, 16, 15, 35, 41, 42, 43, 44, 45, 47, 48, 50, 51, 78, 103, 104, 105, 106, 107, 110, 111, 112, 114, 115, 116, 127, 140, 141, 142, 143, 144, 145, 146, 147, 151, 152, 153, 166, 167, 168, 169, 171, 173, 174, 177, 178, 179, 180, 181, 182, 183, 184, 185, 186, 187, 188, 194, 205, 209, 210, 211, 212, 213, 214, 215, 216, 217, 218, 219, 220, 221, 222, 223, 225, 226, 227, 228, 229, 230, 232, 233, 234, 235, 236, 237, 238, 239, 240, 241, 242, 243, 250, 252]
```

```python
# read columns from txt by line of file
column_names = []
with open('data/colunas.txt', 'r') as f:
    # for index, line
    for i, line in enumerate(f):
        # if index is in columns_to_use
        if i in columns_to_use:
            # append line to column_names
            column_names.append(line.strip())
df = df[column_names]
df = df.dropna(axis=1, how='all')
df.head()
```

```python
df.to_pickle('data/saeb_pb_filtered.pkl')
```
